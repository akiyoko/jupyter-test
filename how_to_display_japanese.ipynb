```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
% matplotlib inline

plt.style.use('ggplot')
```

```python
df = pd.DataFrame({u'データ１': [0, 2, 4],
                   u'データ２': [1, 2, 3]})

df.plot(y=[u'データ１', u'データ２'])
plt.xlabel(u'X軸')
plt.ylabel(u'Y軸')
plt.title(u'日本語・テスト')
df
```

```python
# AppleGothic だと「ー」や「・」とかが文字化けしてしまう
font = {'family': 'AppleGothic'}
matplotlib.rc('font', **font)

df.plot(y=[u'データ１', u'データ２'])
plt.xlabel(u'X軸')
plt.ylabel(u'Y軸')
plt.title(u'日本語・テスト')
```
